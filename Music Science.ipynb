### Spotify Metadata Column Descriptions
The kaggle website that I pulled the datasets from included descriptions, but I decided to cross-reference the [official spotify website](https://developer.spotify.com/documentation/web-api/reference/#/operations/get-several-audio-features) to understand what the data was describing about the songs.

__Beats per Mintue (BPM)__

Beats Per Minute - The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.

__Energy__

Energy - Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.

__Danceability__

Danceability - Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.

__Loudness (dB)__

Loudness - The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typically range between -60 and 0 db.

__Liveness__

Liveness - Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.

__Length (Duration)__

Duration - The duration of the track in seconds.

__Acousticness__

Acousticness - A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.

__Speechiness__

Speechiness - Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.

__Valence__

Valence - A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

__Popularity__

Popularity- The higher the value the more popular the song is.

### How is the correlation with popularity for individual artists?

Knowing all the popular artists, we can see how the makeups of their songs may have correlation with popularity

```python
# creating a list with each artist dataframe
artist_dfs = []
for artist in top_artists:
    artist_dfs.append(spotify2000_df.loc[spotify2000_df['Artist'] == artist])
```

```python
# create dataframe with the artist and popularity
def plot_max_correlations(top_column, value, df, title):
    topColumn_pop = pd.DataFrame({top_column: df[top_column],
                 'popularity': df['Popularity']},
                 columns = [top_column, 'popularity'])
    label_string = 'corr = ' + str(value)
    artist_corr_plot = topColumn_pop.plot(kind='scatter', x=top_column, y='popularity', label=label_string)
    artist_corr_plot.set_title(title + ' - ' + top_column + ' correlation with popularity plot')
    x = topColumn_pop[[top_column]]
    y = topColumn_pop['popularity']
    reg = LinearRegression().fit(x, y)
    ytrain = reg.intercept_ + reg.coef_ * x
    artist_corr_plot.plot(x, y,'ro',x,ytrain,'b-')
```

```python
# looping through each artist dataframe to find the correlation matrix in relation to popularity
# only plotting the correlations > 0.5

for i, artist_df in enumerate(artist_dfs):
    corr_matrix = artist_df.corr()
    top_corr_value = corr_matrix[['Popularity']].sort_values(by='Popularity',ascending=False).values[1][0]
    top_corr_column = corr_matrix[['Popularity']].sort_values(by='Popularity',ascending=False).index[1]
    if top_corr_value > 0.5 and top_corr_column != 'Year' and top_corr_column != 'Index' and top_corr_column != 'Loudness (dB)':
        plot_max_correlations(top_corr_column, top_corr_value, artist_df, top_artists[i])
    
```

It's a lot to unpack, but it's interesting to see that there is better correlation when looking at smaller subsets of the data. Particularly with the artists in this dataset, there are better indiciations of their song's metadata in correlation to popularity. Danceability, valence, and energy seem to be more indicative popularity.

### Visualizing genre averages and comparing

To visualize the genre column averages, I thought it would be a good idea to utilize a radar chart. In this radar chart, I plotted 6 categories that I thought were most important (and for the reason that including other columns like BPM would offset the scale and distract from the overall visualization).

```python
# creating a function to find the averages of each category
# returns a list of those averages to be used for radar charts
def find_avgs(df):
    categories =["Energy", "Danceability", "Liveness", "Valence", "Acousticness", "Speechiness"]
    values = []
    for category in categories:
        values.append(df[category].mean())
    return values

# a list of list of dataframe column averages
values = []

# loop through each genre dataframe to append a list of column averages into values
for genre_df in genre_df_list:
    values.append(find_avgs(genre_df))
```

```python
# radar chart usage in references
categories =["Energy", "Danceability", "Liveness", "Valence", "Acousticness", "Speechiness"]
label_loc = np.linspace(0, 2*np.pi, len(categories), endpoint=False)

plt.figure(figsize=(8, 8))
plt.subplot(polar=True)

colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# loop through the values to plot into radar chart
for df_index, genre_values in enumerate(values):
    plt.plot(label_loc, genre_values, label=(genre_name_list[df_index]))
    plt.fill(label_loc, genre_values, alpha=0.25, facecolor=colors[df_index])

lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
plt.title('Genre category comparison', size=20)
plt.legend()
plt.show()
```

Looking at the radar chart, we can see that hiphop seems to be very prevalent, with the highest averages in all but acousticness. Hiphop wasn't the most popular genre, but it is a top contender with a high median and average. It is interesting to see that there seems to be high averages compared to the other genres. However, that connection and relationship isn't as strong when looking at the opposite. Specifically, indie doesn't seem to be the lowest for any of these values.

## Data Analysis - Decades

Okay! We looked at genres, what if we separate the data into decades? Again, how would that affect the popularity? Can we draw any important conclusions from this?

### Which decade is most popular?

```python
# range of values = 1956 - 2019
# splitting the dataset into decades

# create a sorted dataframe based on year
year_sorted_df = spotify2000_df.sort_values(by='Year',ascending=True)

# creating a dictionary with 
    # - key = decade name 
    # - value = decade dataframe
    
decades_df = {}

# initialize the start value
year_range = 1950

# loop through each decade until finished with dataset year upper limit
while year_range != 2020:
    # create the key based on the decade
    df_name = 'decade_' + str(year_range) + 's'
    # create the value / dataframe by locating the range of rows in between the decade values
    df = year_sorted_df.loc[(year_sorted_df['Year'] >= year_range) & (year_sorted_df['Year'] <= year_range+9)]
    decades_df[df_name] = df
    year_range += 10

decades_df.keys()
```

```python
# creating a dataframe with decades and popularity
decades_popularities = pd.DataFrame({
                'decade_1950s': decades_df['decade_1950s']['Popularity'],
                'decade_1960s': decades_df['decade_1960s']['Popularity'],
                'decade_1970s': decades_df['decade_1970s']['Popularity'],
                'decade_1980s': decades_df['decade_1980s']['Popularity'],
                'decade_1990s': decades_df['decade_1990s']['Popularity'],
                'decade_2000s': decades_df['decade_2000s']['Popularity'],
                'decade_2010s': decades_df['decade_2010s']['Popularity']},
                columns = ['decade_1950s', 'decade_1960s', 'decade_1970s', 'decade_1980s', 'decade_1990s', 'decade_2000s', 'decade_2010s'])
```

```python
# creating a boxplot to see the std and averages
decades_boxplot = decades_popularities.boxplot(column = ['decade_1950s', 'decade_1960s', 'decade_1970s', 'decade_1980s', 'decade_1990s', 'decade_2000s', 'decade_2010s'],
                          figsize=(12,12),
                          showmeans=True)

decades_boxplot.set_title('Boxplots of Decade Popularity')
decades_boxplot.set_xlabel("Decades")
decades_boxplot.set_ylabel("Popularity")
```

Looking at the boxplots, there's not a wide difference with the median or averages between the decades.

### Does decade affect the popularity correlation? 

```python
# looping through each decade dataframe to find the correlation matrix in relation to popularity
# plotting the max correlation graphs
for i, (decade, decade_df) in enumerate(decades_df.items()):
    corr_matrix = decade_df.corr()
    top_corr_value = corr_matrix[['Popularity']].sort_values(by='Popularity',ascending=False).values[1][0]
    top_corr_column = corr_matrix[['Popularity']].sort_values(by='Popularity',ascending=False).index[1]
    plot_max_correlations(top_corr_column, top_corr_value, decade_df, decade)
```

## Data Analysis - Lyrics
Finally, one last thing that could be interesting to look at are the song lyrics of the most popular songs! Let's analyze the lyrics and get a polarity score to see if that can tell us anything.

### Do the top / bottom 100 songs have any similarities with the polarity of their lyrics? How do these two groups compare?

```python
# did top 100 only that for the sake of memory + runtime

# function to scrape lyrics from genius
def scrape_lyrics(artist_name, song_name):
    # replacing spaces with hyphens
    artist_name_hyphens = str(artist_name.replace(' ','-')) if ' ' in artist_name else str(artist_name)
    song_name_hyphens = str(song_name.replace(' ','-')) if ' ' in song_name else str(song_name)
    
    # creating link string 
    link = 'https://genius.com/'+ artist_name_hyphens + '-' + song_name_hyphens + '-' + 'lyrics'
    
    # using requests and beautifulsoup to get the html page
    page = requests.get(link)
    html = BeautifulSoup(page.text, 'html.parser')
    
    # parsing through the html to get the lyrics divs
    lyric_lines = html.find_all("div", class_="Lyrics__Container-sc-1ynbvzw-6 jYfhrf") # Lyrics__Container-sc-1ynbvzw-2 jgQsqn
    
    # combining all the lyrics
    lyrics = ''
    for lyric_line in lyric_lines:
        lyrics += (lyric_line.get_text(separator=' '))
    return lyrics
```
